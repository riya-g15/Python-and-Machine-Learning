```python
#Question_2(iv)
#Venue analysis

ground_avg.groupby(["Ground"])["Mat"].sum().sort_values(ascending=False)[:10]

plt.figure(figsize=(10,8))
data = ground_avg.groupby(["Ground"])["Mat"].sum().sort_values(ascending=False)[:10]
print(type(data))
sns.barplot(y=data.index,x=data,orient="h")
plt.title("Top 10 Venues")
plt.xlabel('Grounds')
plt.ylabel('Matches')
plt.show()
```

These are the top 10 grounds, arranged by the number of matches. Most number of matches have been held in Harare Sports Club, Zimbabwe i.e 50.

```python
ground_avg.groupby(["Ground"])["Runs"].sum().sort_values(ascending=False)[:10]

plt.figure(figsize=(10,8))
data = ground_avg.groupby(["Ground"])["Runs"].sum().sort_values(ascending=True)[:10]
print(type(data))
sns.barplot(y=data.index,x=data,orient="h")
plt.title("Top 10 Venues for maximum runs")
plt.xlabel('Grounds')
plt.ylabel('Runs')
plt.show()
```

Greenfield international stadium has the least runs scored , i.e only around 250. It is worst performing in terms of runs.
