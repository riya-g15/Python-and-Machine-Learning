```python
def createtable():
    # connecting to the database
    connection = sqlite3.connect("Fashion1.db")
    
    # cursor
    crsr = connection.cursor()
  
    com1= """CREATE TABLE VENDOR
    (V_ID CHAR(6) CONSTRAINT PK1 PRIMARY KEY,
    NAME VARCHAR(14) CONSTRAINT NN1 NOT NULL,
    NO_OF_PRODUCTS NUMBER(4) CONSTRAINT CHNN1 CHECK (NO_OF_PRODUCTS >=1) NOT NULL,
    ADHAAR_CARD_NO CHAR(12) CONSTRAINT NNU1 NOT NULL UNIQUE,
    DATE_OF_JOINING DATE,
    PINCODE VARCHAR(6) CONSTRAINT NN2 NOT NULL,
    CITY VARCHAR(15) CONSTRAINT NN3 NOT NULL);"""
    crsr.execute(com1) 
    
    # SQL command to create a table in the database DDL

    com2="""CREATE TABLE WAREHOUSE
    (W_ID CHAR(3) CONSTRAINT PK2 PRIMARY KEY,
    YEAR_OF_EST CHAR(4),
    NO_OF_WORKERS NUMBER(3) CONSTRAINT CHNN2 CHECK (NO_OF_WORKERS >=10) NOT NULL,
    PINCODE VARCHAR(6) CONSTRAINT NN4 NOT NULL,
    CITY VARCHAR(15) CONSTRAINT NN5 NOT NULL);"""
    crsr.execute(com2)
    

    com3="""CREATE TABLE PRODUCT
    (P_ID CHAR(6) CONSTRAINT PK3 PRIMARY KEY,
    DATE_OF_UPLOAD DATE CONSTRAINT NN6 NOT NULL,
    GENDER VARCHAR(10) CONSTRAINT CH1 CHECK (GENDER='FEMALE' OR GENDER='MALE' OR GENDER='NEUTRAL'),
    COLOUR VARCHAR(10),
    TYPE VARCHAR(10),
    RENT_PRICE NUMBER(4) CONSTRAINT CHNN3 CHECK (RENT_PRICE >=499) NOT NULL,
    VENDOR_ID CHAR(6) REFERENCES  VENDOR(V_ID) ON DELETE CASCADE CONSTRAINT NN7 NOT NULL , 
    WAREHOUSE_ID CHAR(3) REFERENCES WAREHOUSE(W_ID) ON DELETE CASCADE CONSTRAINT NN8 NOT NULL);"""
    crsr.execute(com3)
    
    # close the connection
    connection.close()
    
createtable()
```

```python
def Update():
    
    # Connecting to sqlite
    conn = sqlite3.connect('Fashion1.db')

    # Creating a cursor object using
    # the cursor() method
    cursor = conn.cursor()

    # Updating
    cursor.execute('''UPDATE VENDOR SET pincode ='201012' where name = "Hrithik";''')
    cursor.execute('''UPDATE WAREHOUSE SET YEAR_OF_EST='1996' WHERE W_ID='102';''')

    conn.commit()
    conn.close()
    
```

```python
def Delete_Data():
  # Connecting to sqlite
  conn = sqlite3.connect('Fashion1.db')

  # Creating a cursor object using
  # the cursor() method
  cursor = conn.cursor()

  # Updating
  cursor.execute('''DELETE FROM PRODUCT WHERE colour="RED";''')

  # Commit your changes in the database
  conn.commit()

  # Closing the connection
  conn.close()
```

```python
def jointable():
    
    # connecting to the database
    connection = sqlite3.connect("Fashion1.db")
    
    #cursor
    crsr = connection.cursor()
    
    print("Values from 2 tables - vendor and product using Joins: \n")
    com="""SELECT P_ID,TYPE,NAME,VENDOR.V_ID from VENDOR,PRODUCT where PRODUCT.VENDOR_ID=VENDOR.V_ID;"""
    crsr.execute(com)
    
    fin=crsr.fetchall()
    
    for i in fin:
        print(i)
        
    connection.close()
    
```

```python
def aggrefun():
    
    # connecting to the database
    connection = sqlite3.connect("Fashion1.db")
    
    #cursor
    crsr = connection.cursor()
    
    #Average number of workers in all the warehouse to get an estimate of the working capacity of each warehouse.
    com="""select round(avg(no_of_workers),1) as Average_Number_of_workers_in_Warehouse from warehouse;"""
    crsr.execute(com)
    
    a=crsr.fetchall()
    print("Average_Number_of_workers_in_Warehouse")
    for i in a:
        print(i[0])
        
    #Total rent_price of all the products to get the current revenue.
    com1="""select Sum(rent_price) as Total_rent_of_all_products FROM product;"""
    crsr.execute(com1)
    
    b=crsr.fetchall()
    print("\nTotal_rent_of_all_products")
    for i in b:
        print(i[0])
    
    #To know the number of products each vendor has uploaded
    com2="""select vendor_id,count(p_id) from product group by vendor_id ;"""
    crsr.execute(com2)
    
    c=crsr.fetchall()
    print("\nNumber of products uploaded by each Vendor")
    for i in c:
        print(i)
    
    
    #To get the details of the product which have the least rent_price. This could be useful for the customers.
    com3="""select * from product WHERE rent_price in (select min(rent_price) from product);"""
    crsr.execute(com3)
    
    d=crsr.fetchall()
    print("\nLeast expensive products")
    for i in d:
        print(i)
    
    #To see the type of clothes available for rent given a certain range.
    com4="""select type,min(rent_price),max(rent_price) from product group by type having max(rent_price)<800;"""
    crsr.execute(com4)
    
    e=crsr.fetchall()
    print("\nType of Products between range 0-800")
    for i in e:
        print(i)
    
    
    
    connection.close()
    
```

```python
aggrefun()
```

```python
def order_by_rent():
    
    # connecting to the database
    connection = sqlite3.connect("Fashion1.db")
    
    #cursor
    crsr = connection.cursor()
    
    #to display in increasing order of rent_price
    
    com="""select p_id,rent_price,type from product order by rent_price DESC ;"""
    
    print("The products ordered by their rent price (descending order) \n")
    crsr.execute(com)
    fin=crsr.fetchall()
    
    for i in fin:
        print(i)
        
    connection.close()
    
```
