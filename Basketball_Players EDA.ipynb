```python
small_df=df[["GP","draft_pick","weight_kg","height_cm"]]

plt.figure(figsize=(15,8))
sns.heatmap(small_df.corr(),annot=True, linewidths=.2,fmt=".2f",)
plt.show()
```

```python
df.groupby("high_school").count()
```

```python
data = df.groupby(["Team"])["GP"].sum().sort_values(ascending=False)

plt.figure(figsize=(15,8))
sns.barplot(y=data.index,x=data,orient="h")
plt.title("Scores per team")
plt.xlabel("Total pointd")
plt.ylabel("Teams")
plt.show()
```
