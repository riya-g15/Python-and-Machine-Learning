```python
# Function to plot histograms
def cc(df):
    nc = df.select_dtypes(include=['number']).columns.tolist()
    dataframe = df[nc]
    
    for i in range(0,len(nc),2):
        if len(nc) > i+1:
            plt.figure(figsize=(10,4))
            plt.subplot(121)
            sns.distplot(dataframe[nc[i]], kde=False)
            plt.subplot(122)            
            sns.distplot(dataframe[nc[i+1]], kde=False)
            plt.tight_layout()
            plt.show()

        else:
            sns.distplot(dataframe[nc[i]], kde=False)
cc(df)
```

These graph suggest that we should use on parametric approach

```python
from scipy.stats.mstats import winsorize

# Function to treat outliers 
def treat_outliers(dataframe):
    cols = list(dataframe)
    for col in cols:
        if col in dataframe.select_dtypes(include=np.number).columns:
            dataframe[col] = winsorize(dataframe[col], limits=[0.05, 0.1],inclusive=(True, True))
    
    return dataframe 
```
