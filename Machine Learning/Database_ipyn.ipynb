```python
import sqlite3
```

```python
def create():
    conn = sqlite3.connect('IndianTeam.db')
    print("Connected with database successfully\n")

    conn.execute('''CREATE TABLE TEAM_INDIA
            (P_ID INT PRIMARY KEY,
            NAME VARCHAR(20) NOT NULL,
            TYPE_OF_PLAYER CHAR(50),
            MATCHES INT NOT NULL,
            INN INT NOT NULL,
            RUNS INT NOT NULL,
            HIGHEST INT NOT NULL,
            BAT_AVG REAL,
            SR REAL NOT NULL,
            HUNDREDS INT,
            WKTS INT NOT NULL,
            ECON REAL NOT NULL,
            BOWL_AVG REAL);''')
    
    print("Table created\n")

    conn.close()

create()
```

```python
def experience():
    conn = sqlite3.connect('Indianteam.db')

    ans = conn.execute("SELECT NAME,MATCHES,TYPE_OF_PLAYER FROM TEAM_INDIA WHERE MATCHES >= 10 ORDER BY MATCHES DESC;")
  
    for r in ans:
        print(r,'\n')

    conn.execute("CREATE TABLE exp_players AS SELECT * FROM TEAM_INDIA WHERE MATCHES >= 10;")

    conn.commit()
    conn.close()

print("\nPlayers Having Adequate International Experience are as follows  : \n ")
experience()
```

```python
def display_top_bat():
    conn = sqlite3.connect('Indianteam.db')
    
    ans = conn.execute("SELECT NAME,MATCHES,SR,BAT_AVG FROM EXP_PLAYERS WHERE TYPE_OF_PLAYER=\'BATSMAN\' order by SR DESC LIMIT 3;")
    for a in ans:
        print(a)
    print("These are the top 3 batsmen. Do you wish to see top 4 to reconsider?")
    s=input("Y/N")
    while True:
        if(s=='Y'):
            ans = conn.execute("SELECT NAME,MATCHES,SR,BAT_AVG FROM EXP_PLAYERS WHERE TYPE_OF_PLAYER=\'BATSMAN\' order by SR DESC LIMIT 4;")
            for a in ans:
                print(a)
            break
        elif(s=='N'):
            print("Okay! You can select these top 3 for your team!")
            break
        else:
            print("Enter the correct option!")
            continue
        
        
    conn.close()
```

```python
def spinner():
    
    conn = sqlite3.connect('Indianteam.db')
    print("Select the number of top spinners you wish to see: 1/2")
    while True:
        try:
            s=int(input("1/2"))
            if(s==1):
                sp = conn.execute('SELECT * FROM exp_players WHERE TYPE_OF_PLAYER=\'SPINNER\' and ECON < 8.5 ORDER BY WKTS DESC LIMIT 1;')
                for a in sp:
                    print(a)
                print("This is your top Spinner")
                break
            elif(s==2):
                sp = conn.execute('SELECT * FROM exp_players WHERE TYPE_OF_PLAYER=\'SPINNER\' and ECON < 8.5 ORDER BY WKTS DESC LIMIT 2;')
                for a in sp:
                    print(a)
                print("\nWe only have 1 experienced spinner! Sorry!")
                break
            else:
                print("Enter correct choice")
                continue
        except ValueError:
            print("Enter an integer only")
            continue

    conn.close()
```
