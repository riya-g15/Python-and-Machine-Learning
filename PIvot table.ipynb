```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("D:/Downloads/pivotfood.xlsx")
df.head(12)
#df.describe()
```

```python
###Pivoting the data
pd.pivot_table(df,index=["Product"])
```

```python
pd.pivot_table(df,index=["Product","City"])
```

```python
pd.pivot_table(df,index=["Product","City"],values=["TotalPrice"],aggfunc=np.sum)
```

```python
#product wise price total
pd.pivot_table(df,index=["Product"],values=["TotalPrice"],aggfunc=np.sum)
```

```python
#city wise price total
pd.pivot_table(df,index=["City"],values=["TotalPrice"],aggfunc=np.sum)
```

```python
####Quantity per region per city of each product
pd.pivot_table(df,index=["Product","Region"],values=["Quantity"],
columns=["City"],aggfunc=[np.sum],fill_value=0)
```
